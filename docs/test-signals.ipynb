```python
import numpy as np
from conic_tools.analysis.signals import SpikeList
```

```python
# generate synthetic spikes
n_neurons = 100
spks_per_neuron = 20
frequency = 10

spk_times = []
spk_ids = []
for neuron in range(n_neurons):
    isi = np.random.poisson(frequency, spks_per_neuron)
    spks = np.cumsum(isi)
    for i in spks:
        spk_times.append((neuron+1, i))
    spk_ids.append(neuron+1) 
```

```python
sl = SpikeList(spk_times, spk_ids)
```

```python
sl.raster_plot(with_rate=True)
```
